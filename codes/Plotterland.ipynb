```python
def HarryPlotter_dvsJ(N):
    """It's magick as Fuck, it can even plot your soul (but is not a psychotherapist)"""

    fig, ax = plt.subplots(figsize=(5,4), layout='constrained')
    #ax2 = ax.inset_axes([0.55, 0.05, 0.4, 0.4])
    colors = [None, 'midnightblue', 'rebeccapurple', 'purple', 'crimson', 'coral']

    for i in range(1, 6):
        a = np.genfromtxt(f'presentation/distVSJ{N}_{i}.txt', delimiter=',')
        a2 = np.genfromtxt(f'presentation/support_vector{N}_{i}.txt', delimiter=',')
        J = a[:, 0]
        d = a[:, 1]
        J_sv = a2[:, 0]
        d_sv = a2[:, 1]
        ax.plot(J, d, c=colors[i], label=f'noise{i}')
        ax.scatter(np.round(J_sv, 2), np.round(d_sv, 1), facecolor='None', edgecolor=colors[i], s=40)
        #ax2.plot(J, d, c=colors[i], label=f'noise{i}')

    ax.set_xlabel("J", size=15)
    ax.set_ylabel("distance", size=15)
    ax.set_title(f'd V.s. J for N = {N}')

    #ax.text((Jc + 0.1), -0.2, r'$J_{c} =$' + str(round(Jc, 3)))
    ax.axvspan(0.25, 0.35, alpha=0.5, color='gainsboro')
    ax.axvspan(1.65, 1.75, alpha=0.5, color='gainsboro')
    ax.text(0.26, 0.1, r'$\delta J$', c='dimgray', size=15)
    ax.text(1.66, 0.1, r'$\delta J$', c='dimgray', size=15)
    ax.plot(J, np.zeros(len(J)), c='dimgray', linestyle=':')
    

    #ax2 = ax.inset_axes([0.55, 0.1, 0.1, 0.1])
    # sub region of the original image
    #x1, x2, y1, y2 = 0.8, 1.0, -0.25, 0.25
    #ax2.set_xlim(x1, x2)
    #ax2.set_ylim(y1, y2)
    #ax2.set_xticklabels([])
    #ax2.set_yticklabels([])

    #ax.indicate_inset_zoom(ax2, edgecolor="black")

    plt.legend()
    #plt.savefig(f'results/noiseless/dVSJ{N}_noisy.pdf', bbox_inches='tight')

HarryPlotter_dvsJ(6)
```
